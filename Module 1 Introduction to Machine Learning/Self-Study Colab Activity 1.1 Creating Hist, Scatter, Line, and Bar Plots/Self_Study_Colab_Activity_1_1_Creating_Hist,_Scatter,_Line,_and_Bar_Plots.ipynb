```python
import pandas as pd
```

```python
import os
os.makedirs("plots", exist_ok=True)
```

```python
df = pd.read_csv('Self-Study Colab Activity 1.1: Creating Hist, Scatter, Line, and Bar Plots/nba.csv')
```

```python
df.head()
```

```python
#Part One: Histogram
```

```python
ax = df['Age'].plot.hist(bins=10, figsize=(8, 6), title="Histogram of Age")
ax.figure.savefig("Self-Study Colab Activity 1.1: Creating Hist, Scatter, Line, and Bar Plots/plots/hist_age.png")
```

```python
#Part Two: Scatter Plot
```

```python
scat = df.plot(kind='scatter', x='Age', y='Salary', title='Age vs Salary')
scat.figure.savefig("Self-Study Colab Activity 1.1: Creating Hist, Scatter, Line, and Bar Plots/plots/scatter_age_salary.png")
print(scat)
```

```python
#Part Three: Line Plot
```

```python
teamdata = df.filter(items=['Team', 'Name', 'Salary'])
teamdata = teamdata.loc[df.Team == 'Golden State Warriors', 'Name':'Salary']
teamdata = teamdata.set_index('Name')

lines = teamdata['Salary'].plot(kind='line', title='Golden State Warriors Salaries')
lines.figure.savefig("Self-Study Colab Activity 1.1: Creating Hist, Scatter, Line, and Bar Plots/plots/line_salary_gsw.png")
print(lines)
```

```python
#Part Four: Bar Chart
```

```python
bar = teamdata['Salary'].plot(kind='bar', title='Golden State Warriors Salaries (Bar)')
bar.figure.savefig("Self-Study Colab Activity 1.1: Creating Hist, Scatter, Line, and Bar Plots/plots/bar_salary_gsw.png")
print(bar)
```
